```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")


# Task 1: Load and Explore Dataset



data = {
    "CustomerID": range(1, 21),
    "Region": ["North", "South", "East", "West"] * 5,
    "AccountBalance": np.random.randint(1000, 10000, size=20),
    "Transactions": np.random.randint(10, 100, size=20),
    "CreditScore": np.random.randint(300, 850, size=20),
    "Age": np.random.randint(18, 70, size=20)
}

df = pd.DataFrame(data)

try:
    print("✅ Dataset created successfully!")
except Exception as e:
    print("❌ Error:", e)

# Show first rows
print("\nFirst 5 rows:")
print(df.head())

# Info and missing values
print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

# Task 2: Basic Data Analysis


print("\nSummary Statistics:")
print(df.describe())

# Group: average balance by region
avg_balance = df.groupby("Region")["AccountBalance"].mean()
print("\nAverage Account Balance by Region:")
print(avg_balance)

# Group: average transactions by region
avg_txn = df.groupby("Region")["Transactions"].mean()
print("\nAverage Transactions by Region:")
print(avg_txn)


# Task 3: Data Visualization


# 1. Line Chart – Account balance trend
plt.figure(figsize=(8,5))
plt.plot(df["CustomerID"], df["AccountBalance"], marker="o", color="blue", label="Balance")
plt.title("Customer Account Balance Trend")
plt.xlabel("Customer ID")
plt.ylabel("Account Balance ($)")
plt.legend()
plt.show()

# 2. Bar Chart – Average account balance per region
plt.figure(figsize=(8,5))
sns.barplot(x="Region", y="AccountBalance", data=df, palette="Set2", ci=None)
plt.title("Average Account Balance by Region")
plt.xlabel("Region")
plt.ylabel("Average Balance ($)")
plt.show()

# 3. Histogram – Credit Score distribution
plt.figure(figsize=(8,5))
plt.hist(df["CreditScore"], bins=10, color="green", alpha=0.7, edgecolor="black")
plt.title("Distribution of Credit Scores")
plt.xlabel("Credit Score")
plt.ylabel("Number of Customers")
plt.show()

# 4. Scatter Plot – Age vs Account Balance
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="AccountBalance", hue="Region", data=df, palette="Set1")
plt.title("Age vs Account Balance by Region")
plt.xlabel("Customer Age")
plt.ylabel("Account Balance ($)")
plt.legend(title="Region")
plt.show()


# Observations

# - Data shows variation in balances across regions.
# - Some regions (e.g., North/South) may have higher average balances.
# - Credit score distribution spans from low (subprime) to high (excellent).
# - Positive relationship between age and account balance for many customers.
```
